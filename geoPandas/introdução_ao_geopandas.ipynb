### Introdução ao GeoPandas

```python
# Bibliotecas padrão do geopandas
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
# ler os dados com geopandas
data = gpd.read_file('dados/SP_Municipios_2019.shp')
data
```

```python
type(data)
```

```python
# Geopandas utiliza muitas coisas da biblioteca pandas
data.head(10)
```

```python
# Visualização do geoDataFrame
data.plot()
```

```python
# Filtragem na visualização do geoDataFrame
data.plot(figsize=(16,14),facecolor='white',edgecolor='black')
plt.show()
```

```python
# Pegando colunos e linhas especifícas
gdf_sp = data[data['NM_MUN'] == 'São Paulo']
gdf_sp
```

```python
# Editando somente a cidade de são paulo
gdf_sp.plot(figsize=(8,10), facecolor='white',edgecolor='black')
plt.show()
```

```python
# Salvar o geodataframe com formato geojson
filename = 'dados/capital_são_paulo.json'
gdf_sp.to_file(filename, driver='GeoJSON')
```

```python
# Lendo o arquivo .json (GeoJSON) e atribuindo a outra variável
gdf_sp2 = gpd.read_file(filename,driver='GeoJSON')
gdf_sp2
```
